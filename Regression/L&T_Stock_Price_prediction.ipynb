# DATA PRE-PROCESSING

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("LT.csv")
x = dataset.iloc[1:, 1:5].values
y = dataset.iloc[1:, -1].values
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer.fit(x[:, :-1])
x[:, :-1] = imputer.transform(x[:, :-1])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2)
```

# LINEAR REGRESSION MULTIVARIABLE - RESULT


```python
from sklearn.linear_model import LinearRegression
lr_m =  LinearRegression()
lr_m.fit(xtrain,ytrain)
lr_m_pred = lr_m.predict(xtest)
```

```python
test_pred = [[1551.50, 1531.25, 1554, 1525]]
```

```python
print(lr_m.predict(test_pred))
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((ytest.reshape(len(ytest),1), lr_m.predict(xtest).reshape(len(lr_m.predict(xtest)),1)),1))
```

```python
fig = plt.figure(figsize=(10, 6))
plt.plot(lr_m.predict(xtest), color = "red")
plt.plot(ytest, color = "blue")
plt.show()
```

# LINEAR REGRESSION SINGLE VARIABLE - RESULT

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain[:,0].reshape(-1,1), ytrain)
```

```python
lr_pred = lr.predict(xtest[:,0].reshape(-1,1))
print(np.concatenate((ytest.reshape(-1,1), lr_pred.reshape(-1,1)),-1))
```
